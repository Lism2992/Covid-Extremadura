```python
#Importamos librer√≠as.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
#Importamos los datos.

Data_path = "C:/Users/Lism_/Desktop/PractiPyhton/Coronas.xlsx"
Corona_data = pd.read_excel(Data_path)
Corona_data
```

```python
Corona_data.shape
```

```python
#Comenzamos a trasponer para ir montando la cosa.

Data_trans = Corona_data.transpose()
Data_trans

Data_buena = Data_trans.iloc[0:-4, 0:18:2]
Data_buena.index
```

```python
Data_buena.shape
```

```python
Data_buena
```

```python
Data_buena.columns = Data_buena.iloc[0]

Data_buena.iloc[0,:]
Data_buena.loc["Unnamed: 0",:]

Data_buena2 = Data_buena.drop(["Unnamed: 0", "Unnamed: 1"], axis = 0)
Data_buena2
```

```python
Data_buena2 = Data_buena2.astype(int)

Data_buena2.dtypes
```

```python
Largo = len(Data_buena2)
Largo


base = dt.date.today()
date_list = [base - dt.timedelta(days=x) for x in range(Largo)]

date_list[1]
```

```python
date_string = []

for fech in date_list:
    
    now = fech # current date and time

    year = now.strftime("%Y")

    month = now.strftime("%m")


    day = now.strftime("%d")


    time = now.strftime("%H:%M:%S")
    
    date_stri = (year + "/" + month +"/"+ day)
    


    date_str = now.strftime("%m/%d/%Y")
    
    date_string.append(date_stri)
    


    
type(date_string[1])
```

```python
Data_buena2.index = date_string

Data_buena2.index.dtype
Data_buena2.index
```

```python
plt.figure(figsize=(14,25))

# Add title
plt.title("Contagios acumulados para √°reas sanitarias en Extremadura")

sns.heatmap(data=Data_buena2, annot=True, fmt = "5" )
```

```python
# Gr√°fico de tendencia.

plt.figure(figsize=(28,14))

# Add title
plt.title("Contagios acumulados por √°reas sanitarias en Extremadura")

#sns.lineplot(data = Data_buena2, x = Data_buena2.index, y ="EXTREMADURA")
sns.lineplot(data = Data_buena2, dashes = False)
```

```python
# Podemos intentar calcular las tasa de contagio media (media de los contagios que cada persona produce).

Data_buena2.columns[0]

        
        
```

```python
longrow = len(Data_buena2.index)
longcol = len(Data_buena2.columns)

print(longrow)
print(longcol)

contagios_df = pd.DataFrame ()

for zona in range(longcol):
    contagios = []
    nombrecol = Data_buena2.columns[zona]
    for fecha in range(longrow-4):
        
        actual = Data_buena2.iloc[fecha,zona]
        pasada = Data_buena2.iloc[fecha+4,zona]
        
        tasa = (actual-pasada)/pasada
        
        contagios.append(tasa)
        
    contagios_df.insert(zona,nombrecol, contagios) 
        

contagios_df
        

        
        
        
```

```python
plt.figure(figsize=(14,25))

# Add title
plt.title("TC4 para √°reas sanitarias en Extremadura")

sns.heatmap(data=contagios_df, annot=True, fmt = ".3")
```

```python
#Potenciales contagiadores por zona.

longrow = len(Data_buena2.index)
longcol = len(Data_buena2.columns)

diascont = 14

print(longrow)
print(longcol)

contagiadores_df = pd.DataFrame ()

for zona in range(longcol):
    contagiadores = []
    nombrecol = Data_buena2.columns[zona]
    for fecha in range(longrow-diascont):
        
        actual = Data_buena2.iloc[fecha,zona]
        pasada = Data_buena2.iloc[fecha+diascont,zona]
        
        contagiadores_num = (actual-pasada)

        
        contagiadores.append(contagiadores_num)
        
        
    contagiadores_df.insert(zona,nombrecol, contagiadores) 

        

contagiadores_df
```

```python
contagiadores_df.index = date_string[0:-diascont]

contagiadores_df.index.dtype
contagiadores_df.index
```

```python
# Gr√°fico de tendencia.

plt.figure(figsize=(28,14))

# Add title
plt.title("Contagios acumulados por √°reas sanitarias en Extremadura")

#sns.lineplot(data = Data_buena2, x = Data_buena2.index, y ="EXTREMADURA")
sns.lineplot(data = contagiadores_df, dashes = False)
```

```python
contagiadores2_df = contagiadores_df.iloc[:,1:]
```

```python
plt.figure(figsize=(14,25))

# Add title
plt.title("Estimaci√≥n de potenciales contagiadores por √°rea sanitaria en Extremadura")

sns.heatmap(data=contagiadores2_df, annot=True, fmt = "5")
```

```python
# Vamos a ver el √≠ndice personal de contagio.

longrow = len(contagiadores_df.index)
longcol = len(contagiadores_df.columns)

print(longrow)
print(longcol)

ipc_df = pd.DataFrame ()

for zona in range(longcol):
    ipc_list = []
    nombrecol = contagiadores_df.columns[zona]
    for fecha in range(longrow-4):
        
        actual = contagiadores_df.iloc[fecha,zona]
        pasada = contagiadores_df.iloc[fecha+4,zona]
        
        if pasada == 0:           
            ipc_num = 0
        else:
            ipc_num = (actual)/(pasada)
        
        
        ipc_list.append(ipc_num)
        
    ipc_df.insert(zona,nombrecol, ipc_list) 
        

ipc_df
```

```python
ipc_df.index = date_string[0:-diascont-4]

ipc_df.index.dtype
ipc_df.index
```

```python
plt.figure(figsize=(14,25))

# Add title
plt.title("R0 estimado sin casos importados por √°rea sanitaria en Extremadura")

sns.heatmap(data=ipc_df, annot=True)
```
