```python
#Importamos librerías.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
#Importamos los datos.

Data_path = "C:/Users/Lism_/Desktop/PractiPyhton/Coronas.xlsx"
Corona_data = pd.read_excel(Data_path)
Corona_data
```

```python
#Comenzamos a trasponer para ir montando la cosa.

Data_trans = Corona_data.transpose()
Data_trans

Data_buena = Data_trans.iloc[0:-4, 1:18:2]
Data_buena.index
```

```python
Data_buena.shape
```

```python
Data_buena.columns = Data_buena.iloc[0]

Data_buena.iloc[0,:]
Data_buena.loc["Unnamed: 0",:]

Data_buena2 = Data_buena.drop(["Unnamed: 0", "Unnamed: 1"], axis = 0)
Data_buena2.head()

# En este punto ya tenemos la estructura montada.
```

```python
Data_buena2.index
```

```python
# Definimos nuestras variables como floats

Data_buena2 = Data_buena2.astype(float)

Data_buena2.dtypes
```

```python
# Vamos a intentar hacer una lista de fechas de la misma longitud que los indices y ponerla como index.


Largo = len(Data_buena2)
Largo


base = dt.date.today()
date_list = [base - dt.timedelta(days=x) for x in range(Largo)]

date_list[1]
```

```python
# Vamos a elaborar un índice que sea una cadena de strings.

date_string = []

for fech in date_list:
    
    now = fech # current date and time

    year = now.strftime("%Y")

    month = now.strftime("%m")


    day = now.strftime("%d")


    time = now.strftime("%H:%M:%S")
    
    date_stri = (year + "/" + month +"/"+ day)
    
    print(date_stri)


    date_str = now.strftime("%m/%d/%Y")
    
    date_string.append(date_stri)
    


    
type(date_string[1])
```

```python
#nditex = list(range(len(Data_buena2)))
#inditex

#Data_buena3 = Data_buena2
Data_buena2.index = date_string

Data_buena2.index.dtype
Data_buena2.index
```

```python
# Gráfico de zonas calientes.

plt.figure(figsize=(14,25))

# Add title
plt.title("TMC4 por áreas sanitarias en Extremadura")

sns.heatmap(data=Data_buena2, annot=True, fmt = ".3")
```

```python
### El reporte de máximos.

blankspace = " "

for area in range(0,(len(Data_buena2.columns))):
    
    col = Data_buena2.iloc[:,area]
    
    maxarea = col.max()
    
    Data_buena2.columns
    
    print(Data_buena2.columns[area], "=", maxarea, 10*blankspace, "Fecha del máximo =", col.idxmax())
```
